You need to create a program that calculates the cost of shipping a package based on its weight and destination üì¶üí∞

üåç Domestic shipping within the country :

* Packages weighing up to 1 kg cost Rs.50 to ship ü§ë
Each additional kg costs Rs.10 extra üí∞

üåé International shipping to other countries üåè:

* Packages weighing up to 1 kg cost Rs.500 to ship ü§ë

Each additional kg costs Rs.100 extra üí∞
Ready to ship? üöõüöÄ

```
#function to calculate the shipping cost based on package weight and delivery location

def calculate_shipping_cost(package_wt, delivery_location): #defining function and setting parameters

    if delivery_location == 'Domestic delivery' and package_wt == 1:
        shipping_cost_inr = 50
    elif delivery_location == 'Domestic delivery' and package_wt > 1:
        shipping_cost_inr = 50 + (package_wt - 1) * 10
    elif delivery_location == 'International delivery' and package_wt == 1:
        shipping_cost_inr = 500
    elif delivery_location == 'International delivery' and package_wt > 1:
        shipping_cost_inr = 500 + (package_wt - 1) * 100
    else:
        return "Invalid destination entered. Enter 'Domestic delivery' or 'International delivery'."

    return shipping_cost_inr
```

```
#test the funcion
package_wt = float(input('Enter the wt:'))
delivery_location = input('Enter the location:')
shipping_cost = calculate_shipping_cost(package_wt, delivery_location)
print(f"Shipping cost is: {shipping_cost}")
```

**Instructions:**

1. First, import the "random" module to generate a random number for the game.
2. Next, enter the number of tries you want to have and set a counter to keep track of how many tries you are left with.
3. Now, use a while loop to allow yourself to keep guessing until you either guess the correct number or run out of tries.
4. Finally, if you run out of tries without guessing the correct number, print a message telling what the secret number was.

```
import random
#Building a function that will act as a guessing game which will provide the user a certain number of attempts to guess the number it has generated
#If the user guesses the right number it will print a congratulatory message else a better luck next time message
def Guessing_Game():
    """Plays a number guessing game with the user."""

    secret_number = random.randint(1, 100) #Generates a random number between 1 to 100
    max_tries = 4 #setting the number of attempts for the user
    tries= 0 #setting the initial tries as 0 before the games starts

    print("Welcome to the Guessing Game!")
    print("I'm thinking of a number between 1 and 100.")

   #Using while loop to continue the game till the max_tries limit is not reached
    while tries < max_tries:
        try: #using try and except to handle errors and code crash
            guess = int(input(f"Attempt{tries+1}: Enter your guess: "))
            tries += 1 #Increments the guess counter

        #Giving hints to the user whether his guessed number is larger or smaller than the generated number so that he can make a better guess in the next attempt
            if guess < secret_number:
               print("Too low! Try again.")
            elif guess > secret_number:
               print("Too high! Try again.")
            else:
               print(f"Congratulations! You guessed the number in {tries} attempts.")
               return #Exits the function if the guess is correct

        except ValueError:
            print("Invalid input. Please enter a number.")

    print(f"Sorry, you've run out of attempts. The number was {secret_number}.") #Reveals the secret number if the user fails

#Start the guessing game
Guessing_Game()
```

## **Building a Fibonacci Sequence Generator**

In this activity, you will build a program that generates a Fibonacci sequence using conditional looping in Python. The Fibonacci sequence is a series of numbers in which each number is the sum of the two preceding numbers. The sequence starts with 0 and 1, and the next number in the sequence is the sum of the previous two numbers.

```
#creating a fibonacci series function using while condition loop
def fibonacci_sequence_generator(n): # n = number of terms that the user enters as input for the sequence

   sequence = [1, 1]
#using while loop to continuously adding the last 2 terms of the sequence and appending it to the sequence till the total number of terms matches the 'n' terms
   while len(sequence) < n:
      next_term = sequence[-1] + sequence[-2]
      sequence.append(next_term)
   return sequence


num_of_terms = int(input(f"Enter the number of terms: "))

fibonacci_numbers = fibonacci_sequence_generator(num_of_terms)

print(f"Fibonacci sequence is: {fibonacci_numbers}")
```

## **Counting the Digit**

You have been given a task to write a Python program that counts the number of times a specific digit appears in a range of numbers. You need to use a for loop to iterate over the range of numbers, and a while loop to check each digit in the number.

Example:

* Enter start number: 100
* Enter end number: 150
* Enter digit to count: 5
* The digit 5 appears 6 times between 100 and 150.


```
#Building a digit counter that will count the number of times the digit which the user has entered, gets repeated in the range entered by the user

def digit_counter(start_num, end_num, single_digit_input):

  count = 0 #initial count

  #looping through the range of numbers entered by the user using for loop
  for num in range(start_num, end_num + 1):
    temp_num = num
    while temp_num > 0: #using while loop to perform conditional calculation
       if temp_num % 10 == single_digit_input:
         count += 1
       temp_num //= 10
  return count


start_num = int(input("Enter start number: "))
end_num = int(input("Enter end number: "))
single_digit_input = int(input("Enter a single digit: "))

digit_count = digit_counter(start_num, end_num, single_digit_input)

print(f"The digit {single_digit_input} appears {digit_count} times between {start_num} and {end_num}")
```

**Instructions:**

1. First, input two numbers as the range (start and end) between which the program should count the occurrence of a digit.

## **Building a Simple Calculator** üßÆüî¢


In this activity, you will build a simple calculator program in Python using Object-Oriented Programming (OOP) concepts.

```
# Define a simple calculator class
class Calculator:
  #building the _init_ constructor
  def __init__(self, num1, num2):
    self.num1 = num1
    self.num2 = num2
  #building the add method under the Calculator class
  def add(self):
    result= self.num1 + self.num2
    return result
  #building the substract method under the Calculator class
  def substract(self):
    result= self.num1 - self.num2
    return result
  #building the multiply method under the Calculator class
  def multiply(self):
    result= self.num1 * self.num2
    return result
  #building the divide method under the Calculator class
  def divide(self):
    result= self.num1 / self.num2 if self.num2 != 0 else 'Cannot divide by zero.'
    return result
  #building the calculate method under the Calculator class
  def calculate(self, operation):
    try:
      if operation == '+':
       result = self.add()
      elif operation == '-':
       result = self.substract()
      elif operation == '*':
       result = self.multiply()
      elif operation == '/':
       result = self.divide()
      else:
       result = "Invalid Operation"
      return result
    except (ValueError, IndexError, ZeroDivisionError):
      return "Invalid Input"
```

```
# Test Case
obj = Calculator(4.6, 2)
print(obj.calculate('/'))
```

üëâ **Instructions:**

1. Start by defining a class called Calculator.
2. Define four methods within the class: add, subtract, multiply, and divide.
3. Each method should take two arguments: num1 and num2.
4. The add method should return the sum of num1 and num2.
5. The subtract method should return the difference between num1 and num2.
6. The multiply method should return the product of num1 and num2.
7. The divide method should return the result of dividing num1 by num2. However, if num2 is zero, the method should raise a ValueError with the message "Cannot divide by zero".

Create a Bank Account Management System using Object-Oriented Programming (OOPs) in Python. The system should allow users to create and manage bank accounts, deposit and withdraw funds, and view account details.

```
# Define a BankAccount class
class BankAccount:
  #building the _init_ constructor
  def __init__(self, name, account_number, balance):
    self.name = name
    self.account_number = account_number
    self.balance = balance

  #building the deposit method which will add all the deposit amount to the balance and print a message to the customer
  def deposit(self, amount):
    if amount > 0:
      self.balance += amount
      print(f"Deposit successful. New balance: {self.balance}")
    else:
      print("Invalid deposit amount.")

  #building the withdraw method which will substract all the withdrawn amount from the balance and print a message to the customer
  def withdraw(self, amount):
    if amount > 0:
      if self.balance >= amount:
        self.balance -= amount
        print(f"Withdrawal successful. New balance: {self.balance}")
      else:
        print("Insufficient Balance.")
    else:
      print("Invalid withdrawal amount.")

  #building the account_details method which will provide all the account details of the customer
  def account_details(self):
    print(f"Name: {self.name}, ",
          f"Account_number: {self.account_number}, ",
          f"Balance: {self.balance}")
```

## **What's the time?** ‚åö‚åö

Create a class Time which takes two inputs: hours and minutes to instantiate.

* Construct a method DisplayTime() which displays the time in AM/PM formats.

 * For example if the input is 14 hours and 45 mins, then this method will print "The time is 2:45 PM".
 * If the inputted hours exceeds 23 then print the message "The input hours should be less than 24" and if the inputted minutes exceeds 59 then print the message "The input minutes should be less than 60."
 * Also if the input is 12 hours 30 minutes, then the displayed time would be 12:30 PM

* Construct a method DisplayRatio() which should display the ratio of minutes to hours.
 * For example, (8 hours and 16 mins) should display 2. Use try, except block to account for ZeroDivisionError.

time.png

```
class Time:
  def __init__(self, hours, minutes):
    # Validate and instantiate hours and minutes
    if not (0 <= hours <= 24):
      print("The input hours should be less than 24")
      self.hours = None # Set to None to indicate invalid state
      self.minutes = None
      return
    if not (0 <= minutes <= 59):
      print("The input minutes should be less than 60")
      self.hours = None
      self.minutes = None
      return

    self.hours = hours
    self.minutes = minutes

  def DisplayTime(self):
    if self.hours is None or self.minutes is None:
      return # Do nothing if time is invalid

    # Logic to display time in AM/PM format
    if self.hours == 0 and self.minutes in range(0, 10):
        print(f'The time is: 12:0{self.minutes} AM')
    elif self.hours == 0 and self.minutes in range(10, 60):
        print(f'The time is: 12:{self.minutes} AM')
    elif self.hours in range(1, 12) and self.minutes in range(0, 10):
       print(f'The time is: {self.hours}:0{self.minutes} AM')
    elif self.hours in range(1, 12) and self.minutes in range(10, 60):
       print(f'The time is: {self.hours}:{self.minutes} AM')
    elif self.hours == 12 and self.minutes in range(0, 10):
       print(f'The time is: {self.hours}:0{self.minutes} PM')
    elif self.hours == 12 and self.minutes in range(10, 60):
       print(f'The time is: {self.hours}:{self.minutes} PM')
    elif self.hours in range(13, 24) and self.minutes in range(0, 10):
       print(f'The time is: {self.hours-12}:0{self.minutes} PM')
    elif self.hours in range(13, 24) and self.minutes in range(10, 60):
       print(f'The time is: {self.hours-12}:{self.minutes} PM')
    elif self.hours == 24 and self.minutes in range(0, 10):
       print(f'The time is: 12:0{self.minutes} AM') # 24:00 is 12:00 AM
    elif self.hours == 24 and self.minutes in range(10, 60):
       print(f'The time is: 12:{self.minutes} AM') # 24:XX is 12:XX AM

  def DisplayRatio(self):
    try:
      if self.hours is None or self.minutes is None:
        return # Do nothing if time is invalid

      # Calculate the ratio of minutes to hours
      ratio = self.minutes / self.hours
      print(f"The ratio of minutes to hours is: {ratio:.2f}")
    except ZeroDivisionError:
      print("Cannot calculate ratio: Hours cannot be zero for division.")
```
