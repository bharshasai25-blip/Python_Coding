## **Duplicate Songs**

üéµüé∂ You have a list of songs that you like to listen to, but you notice that some of the songs have duplicate names. You want to create a new set that contains the unique names of all the songs and keep the songs whose name doesn't start with "S" and doesn't end with "n" in your list using set comprehension.

Can you write the code to create a new set that contains the unique names of all the songs in your list using set comprehension only?

üëâ Hint: Set comprehension is similar to list comprehension, but with curly braces {} instead of square brackets [].

```
#list of songs
songs = ['Bohemian Rhapsody', 'Stairway to Heaven', 'Bohemian Rhapsody', 'Hotel California', 'November Rain', 'Stairway to Heaven', 'Sweet Child O\' Mine']

#converting the list to set
songs_set = set(songs)
#performing set comprehension on the set using for loop and if conditional statement to find the specific songs that meet the specific conditions mentioned under if conditions
songs_new_set = {x for x in songs_set if x[0]!='S' and x[-1]!='n'}
print(songs_new_set)
```

```
# Here's an example list of songs
songs = ['Bohemian Rhapsody', 'Stairway to Heaven', 'Bohemian Rhapsody', 'Hotel California', 'November Rain', 'Stairway to Heaven', 'Sweet Child O\' Mine']

songs_set = set(songs)
songs_new_set = {x for x in songs_set if x.startswith('S') and x.endswith('n')}
print(songs_new_set)
```

## **Building a Simple Calculator** üßÆüî¢


In this activity, you will build a simple calculator program in Python using Object-Oriented Programming (OOP) concepts.

```
# Define a simple calculator class
class Calculator:
  #building the _init_ constructor
  def __init__(self, num1, num2):
    self.num1 = num1
    self.num2 = num2
  #building the add method under the Calculator class
  def add(self):
    result= self.num1 + self.num2
    return result
  #building the substract method under the Calculator class
  def substract(self):
    result= self.num1 - self.num2
    return result
  #building the multiply method under the Calculator class
  def multiply(self):
    result= self.num1 * self.num2
    return result
  #building the divide method under the Calculator class
  def divide(self):
    result= self.num1 / self.num2 if self.num2 != 0 else 'Cannot divide by zero.'
    return result
  #building the calculate method under the Calculator class
  def calculate(self, operation):
    try:
      if operation == '+':
       result = self.add()
      elif operation == '-':
       result = self.substract()
      elif operation == '*':
       result = self.multiply()
      elif operation == '/':
       result = self.divide()
      else:
       result = "Invalid Operation"
      return result
    except (ValueError, IndexError, ZeroDivisionError):
      return "Invalid Input"
```

```
# Test Case
obj = Calculator(4.6, 2)
print(obj.calculate('/'))
```

üëâ **Instructions:**

1. Start by defining a class called Calculator.
2. Define four methods within the class: add, subtract, multiply, and divide.
3. Each method should take two arguments: num1 and num2.
4. The add method should return the sum of num1 and num2.
5. The subtract method should return the difference between num1 and num2.
6. The multiply method should return the product of num1 and num2.
7. The divide method should return the result of dividing num1 by num2. However, if num2 is zero, the method should raise a ValueError with the message "Cannot divide by zero".

## üèÄ**Basketball Game Score Tracker**üìà

Create a class that will keep track of the score of a basketball game. The class should have methods for updating the score and printing the current score.

basketball-net.png

```
# Create BasketballGame class
class BasketballGame:
  def __init__(self):
    self.Team1_score= 0
    self.Team2_score= 0

  def update_score(self, team, points):
    """Updates the score for the specified team."""
    if team == 1:
      self.Team1_score += points
    elif team == 2:
      self.Team2_score += points
    else:
      print("Invalid team number. Please use 1 or 2.")

  def print_score(self):
    """Prints the current score of the game."""
    print(f"Current Score: Team 1 = {self.Team1_score}, Team 2 = {self.Team2_score}")
```

```
# Example Usage:
game = BasketballGame()
game.print_score()

game.update_score(1,2)
game.print_score()

game.update_score(2,4)
game.print_score()

game.update_score(1, 1)
game.print_score()
```

1. Start by creating a class called "BasketballGame".

## **Bank Account Management System** üíªüè¶

Create a Bank Account Management System using Object-Oriented Programming (OOPs) in Python. The system should allow users to create and manage bank accounts, deposit and withdraw funds, and view account details.

bank.png

```
# Define a BankAccount class
class BankAccount:
  #building the _init_ constructor
  def __init__(self, name, account_number, balance):
    self.name = name
    self.account_number = account_number
    self.balance = balance

  #building the deposit method which will add all the deposit amount to the balance and print a message to the customer
  def deposit(self, amount):
    if amount > 0:
      self.balance += amount
      print(f"Deposit successful. New balance: {self.balance}")
    else:
      print("Invalid deposit amount.")

  #building the withdraw method which will substract all the withdrawn amount from the balance and print a message to the customer
  def withdraw(self, amount):
    if amount > 0:
      if self.balance >= amount:
        self.balance -= amount
        print(f"Withdrawal successful. New balance: {self.balance}")
      else:
        print("Insufficient Balance.")
    else:
      print("Invalid withdrawal amount.")

  #building the account_details method which will provide all the account details of the customer
  def account_details(self):
    print(f"Name: {self.name}, ",
          f"Account_number: {self.account_number}, ",
          f"Balance: {self.balance}")
```

```
# Test Case
Customer1 = BankAccount('Alex', 'AX230005678', 50000)
Customer1.deposit(20000)
Customer1.withdraw(80000)
Customer1.account_details()
```

üëâ **Instructions:**

1. Define a BankAccount class with an init() method that takes the name, account_number, and balance as parameters and initializes the instance variables.

2. Define the deposit() method that takes the amount to be deposited as a parameter and adds it to the balance. Also, print a message with the new balance.

## **What's the time?** ‚åö‚åö

Create a class Time which takes two inputs: hours and minutes to instantiate.

* Construct a method DisplayTime() which displays the time in AM/PM formats.

 * For example if the input is 14 hours and 45 mins, then this method will print "The time is 2:45 PM".
 * If the inputted hours exceeds 23 then print the message "The input hours should be less than 24" and if the inputted minutes exceeds 59 then print the message "The input minutes should be less than 60."
 * Also if the input is 12 hours 30 minutes, then the displayed time would be 12:30 PM

* Construct a method DisplayRatio() which should display the ratio of minutes to hours.
 * For example, (8 hours and 16 mins) should display 2. Use try, except block to account for ZeroDivisionError.

time.png

```
class Time:
  def __init__(self, hours, minutes):
    # Validate and instantiate hours and minutes
    if not (0 <= hours <= 24):
      print("The input hours should be less than 24")
      self.hours = None # Set to None to indicate invalid state
      self.minutes = None
      return
    if not (0 <= minutes <= 59):
      print("The input minutes should be less than 60")
      self.hours = None
      self.minutes = None
      return

    self.hours = hours
    self.minutes = minutes

  def DisplayTime(self):
    if self.hours is None or self.minutes is None:
      return # Do nothing if time is invalid

    # Logic to display time in AM/PM format
    if self.hours == 0 and self.minutes in range(0, 10):
        print(f'The time is: 12:0{self.minutes} AM')
    elif self.hours == 0 and self.minutes in range(10, 60):
        print(f'The time is: 12:{self.minutes} AM')
    elif self.hours in range(1, 12) and self.minutes in range(0, 10):
       print(f'The time is: {self.hours}:0{self.minutes} AM')
    elif self.hours in range(1, 12) and self.minutes in range(10, 60):
       print(f'The time is: {self.hours}:{self.minutes} AM')
    elif self.hours == 12 and self.minutes in range(0, 10):
       print(f'The time is: {self.hours}:0{self.minutes} PM')
    elif self.hours == 12 and self.minutes in range(10, 60):
       print(f'The time is: {self.hours}:{self.minutes} PM')
    elif self.hours in range(13, 24) and self.minutes in range(0, 10):
       print(f'The time is: {self.hours-12}:0{self.minutes} PM')
    elif self.hours in range(13, 24) and self.minutes in range(10, 60):
       print(f'The time is: {self.hours-12}:{self.minutes} PM')
    elif self.hours == 24 and self.minutes in range(0, 10):
       print(f'The time is: 12:0{self.minutes} AM') # 24:00 is 12:00 AM
    elif self.hours == 24 and self.minutes in range(10, 60):
       print(f'The time is: 12:{self.minutes} AM') # 24:XX is 12:XX AM

  def DisplayRatio(self):
    try:
      if self.hours is None or self.minutes is None:
        return # Do nothing if time is invalid

      # Calculate the ratio of minutes to hours
      ratio = self.minutes / self.hours
      print(f"The ratio of minutes to hours is: {ratio:.2f}")
    except ZeroDivisionError:
      print("Cannot calculate ratio: Hours cannot be zero for division.")
```
