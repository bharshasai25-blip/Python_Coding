ðŸŒ Domestic shipping within the country :

* Packages weighing up to 1 kg cost Rs.50 to ship ðŸ¤‘
Each additional kg costs Rs.10 extra ðŸ’°

ðŸŒŽ International shipping to other countries ðŸŒ:

* Packages weighing up to 1 kg cost Rs.500 to ship ðŸ¤‘

Each additional kg costs Rs.100 extra ðŸ’°
Ready to ship? ðŸš›ðŸš€

JOP.gif

```
#function to calculate the shipping cost based on package weight and delivery location

def calculate_shipping_cost(package_wt, delivery_location): #defining function and setting parameters

    if delivery_location == 'Domestic delivery' and package_wt == 1:
        shipping_cost_inr = 50
    elif delivery_location == 'Domestic delivery' and package_wt > 1:
        shipping_cost_inr = 50 + (package_wt - 1) * 10
    elif delivery_location == 'International delivery' and package_wt == 1:
        shipping_cost_inr = 500
    elif delivery_location == 'International delivery' and package_wt > 1:
        shipping_cost_inr = 500 + (package_wt - 1) * 100
    else:
        return "Invalid destination entered. Enter 'Domestic delivery' or 'International delivery'."

    return shipping_cost_inr
```

```
#test the funcion
package_wt = float(input('Enter the wt:'))
delivery_location = input('Enter the location:')
shipping_cost = calculate_shipping_cost(package_wt, delivery_location)
print(f"Shipping cost is: {shipping_cost}")
```

## You and your friends are playing a board game where you roll a dice and move your game piece accordingly. However, you want to know the average number rolled by each player to see who's really the luckiest. Write a program that takes a list of rolls as input for each player and uses a for loop to calculate the average roll for each player. The winner gets bragging rights for being the luckiest!




dice-gif.gif

```
#list of rolls of all players
player_rolls = [
    [3, 5, 6, 2, 1],  # player 1's rolls
    [4, 4, 6, 3, 2],  # player 2's rolls
    [1, 5, 6, 6, 4]   # player 3's rolls
] # list of a list
```

```
max_avg = 0
max_index = 0

#looping through all the player_rolls to find the average roll of each player
for index in range(len(player_rolls)):
  print(player_rolls[index])
  average_index = sum(player_rolls[index])/len(player_rolls[index])  #calculating the average roll of each player
  print(average_index)
  if average_index>max_avg:                                          #setting condition for the maximum average roll
    max_avg = average_index                                          #finding the maximum average roll
    max_index = index                                                #finding the index of the player with the highest average rolls
print(f"The winner is player {max_index+1} with an average roll of {max_avg}")
```

**Instructions:**

1. Define a list of rolls for each player, where each list contains the numbers rolled by the player.
2. Use a for loop to iterate over each player's rolls in the list of player rolls.
3. For each player, calculate the average roll by summing up all the rolls and dividing by the number of rolls. You can use the formula sum(rolls) / len(rolls) to do this.
4. Print out each player's average roll using f-strings to display the player number and their average roll to two decimal places.

## Write a Python program that analyzes a list of names and sorts them into two categories: those that start with a vowel and those that start with a consonant. To achieve this, you'll use a nested loop and some basic conditional statements.

```
#list of names
names = ["Alice", "Bob", "Eve", "Charlie", "Ivy", "David", "Olivia", "Peter"]

#looping through the list to find whether a name starts with a vowel or a consonent
for name in names:
  if name.startswith("A") or name.startswith("E") or name.startswith("I") or name.startswith("O") or name.startswith("U"):
    print(f"{name} starts with a vowel")
  else:
    print(f"{name} starts with a consonant")
```
